```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
def getData(link):
  try:
    raw = requests.get(link)
    raw.raise_for_status()
    data = BeautifulSoup(raw.text,'html5lib')
  except Exception as e:
    print(e)
  return data
```

```python
data = getData('https://www.goodreads.com/shelf/show/book-reviews')
bookElements = data.find_all('div', class_='elementList')
listOfBooks = []

for book in bookElements:
  bookDetail = book.find('div', class_='left')
  if bookDetail is None:
    break
  bookLinkTag = bookDetail.find('a',{'class':'bookTitle'})
  bookAuthor = bookDetail.find('a',{'class':'authorName'}).span.text
  bookLink = bookLinkTag['href']
  baseDict = {
      'bookLink': 'https://www.goodreads.com' + bookLink,
      'bookTitle': bookLinkTag.text,
      'bookAuthor': bookAuthor}
  listOfBooks.append(baseDict)
```

```python
for book in listOfBooks:
  data = getData(book['bookLink'])
  bookRating = data.find('div', class_='RatingStatistics__rating').text
  noOfRatings = data.find('span', {'data-testid':'ratingsCount'}).text.split()[0]
  noOfReviews = data.find('span', {'data-testid':'reviewsCount'}).text.split()[0]
  book['bookRating'] = bookRating
  book['noOfRatings'] = noOfRatings
  book['noOfReviews'] = noOfReviews
  genres = data.find('div', {'data-testid':"genresList"}).find_all('span', class_='BookPageMetadataSection__genreButton')
  genre = [i.text for i in genres]
  book['genres'] = genre
```

```python
import pandas as pd

df = pd.DataFrame(listOfBooks)
df.to_csv('books.csv',sep=',',encoding='utf-8')
```
